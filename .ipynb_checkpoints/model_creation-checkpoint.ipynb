## import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.preprocessing import image_dataset_from_directory
```

```python
physical_devices = tf.config.list_physical_devices('GPU')

print(physical_devices)

if len(physical_devices) >= 1:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

## load training images

#### 80% data for training, 20% data for validation

```python
data_dir = "/content/drive/MyDrive/HPE_Proj/food_recognition_proj/images/train"
## if running in local, then use the path in next line
## data_dir = "images/train"

BATCH_SIZE = 16

IMG_SIZE = (256, 256)

train_dataset = image_dataset_from_directory(data_dir,
                                             validation_split=0.2,
                                             subset='training',
                                             seed = 123,
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE)
```

```python
val_dataset = image_dataset_from_directory(data_dir,
                                             validation_split=0.2,
                                             subset='validation',
                                             seed = 123,
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE)
```

```python
class_names = train_dataset.class_names

print(class_names)
```

### show the first 9 images and labels from training set

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for images, labels in train_dataset.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```

### configure dataset for performance

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
val_dataset = val_dataset.prefetch(buffer_size=AUTOTUNE)
```

### apply data augmentation

```python
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.experimental.preprocessing.RandomFlip('horizontal_and_vertical'),
  tf.keras.layers.experimental.preprocessing.RandomRotation(0.2),
  tf.keras.layers.experimental.preprocessing.RandomContrast(0.2),
  tf.keras.layers.experimental.preprocessing.RandomZoom(height_factor=(-0.2, 0.2), width_factor =(-0.2, 0.2), fill_mode='constant'),
  tf.keras.layers.experimental.preprocessing.RandomTranslation(height_factor=(-0.2, 0.3), width_factor=(-0.2, 0.3), fill_mode='reflect'),
])
```

### check data augmentation sample result

```python
for image, _ in train_dataset.take(1):
    plt.figure(figsize=(10, 10))
    first_image = image[0]
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        augmented_image = data_augmentation(tf.expand_dims(first_image, 0))
        plt.imshow(augmented_image[0] / 255)
        plt.axis('off')
```

## Rescale pixel values

```python
from keras.models import Model
import numpy as np

rescale = tf.keras.layers.experimental.preprocessing.Rescaling(1./127.5, offset= -1)
```

## create base model from pre-trained convnets

```python
def getBaseModel(model_name, fine_tune_at=-10):
    

    # Create the base model from the pre-trained model 
    IMG_SHAPE = IMG_SIZE + (3,)

    base_model = None

    if model_name == "ResNet152V2":
        ## resnet model
        base_model = tf.keras.applications.ResNet152V2(input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
    elif model_name == "Xception":
        ## xception model
        base_model = tf.keras.applications.Xception(input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
    elif model_name == "InceptionResNetV2":
        ## InceptionResNetV2
        base_model = tf.keras.applications.InceptionResNetV2(input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
    elif model_name == "InceptionV3":
        ## InceptionNetV3
        base_model = tf.keras.applications.InceptionV3(input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
    else:
        print("unknown model")
    
    base_model.trainable = True

    # Let's take a look to see how many layers are in the base model
    print("Number of layers in the base model: ", len(base_model.layers))

    # Fine-tune from this layer onwards
    print("Fine tune from layer ", fine_tune_at)

    # Freeze all the layers before the `fine_tune_at` layer
    for layer in base_model.layers[:fine_tune_at]:
        layer.trainable =  False

    #base_model.summary()

    return base_model
```

### comfigure model parameters

```python
def configure_model_parameters(base_model, output_units, activation_function, learning_rate):
    outout_units = len(class_names)
    activation = activation_function

    IMG_SHAPE = IMG_SIZE + (3,)
    inputs = tf.keras.Input(shape=IMG_SHAPE)
    x = data_augmentation(inputs)

    # from (0, 255) to a range (-1., +1.), the normalization layer
    # does the following, outputs = (inputs - mean) / sqrt(var)
    norm_layer = keras.layers.experimental.preprocessing.Normalization()
    mean = np.array([127.5] * 3)
    var = mean ** 2
    # Scale inputs to [-1, +1]
    x = norm_layer(x)
    norm_layer.set_weights([mean, var])

    x = base_model(x)

    global_avg_layer = tf.keras.layers.GlobalAveragePooling2D()
    x = global_avg_layer(x)

    x = tf.keras.layers.Dropout(0.2)(x)
    #x = keras.layers.Flatten()(x)
    #x = keras.layers.Dense(256, activation='relu')(x)
    #x = keras.layers.Dropout(0.5)(x)

    #x = tf.keras.layers.Dense(outout_units, activation=activation, kernel_regularizer=tf.keras.regularizers.l2())(x)
    x = tf.keras.layers.Dense(outout_units, kernel_regularizer=tf.keras.regularizers.l2())(x)
    outputs = tf.keras.layers.Softmax()(x)
    model = tf.keras.Model(inputs, outputs)

    base_learning_rate = learning_rate
    
    model.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
                  loss='sparse_categorical_crossentropy', 
                  metrics=['accuracy'])

    return model

def configure_model(model_name, fine_tune_at, output_units, activation_function, learning_rate):
    the_base_model = getBaseModel(model_name, fine_tune_at)

    final_model = configure_model_parameters(the_base_model, output_units, activation_function, learning_rate)

    return final_model
```

### scheduler and early stop

```python
def scheduler(epoch, lr):
    if epoch % 5 == 0:
        return lr / 2
    else: 
        return lr

st_callback = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=6, restore_best_weights=True)
lr_callback = tf.keras.callbacks.LearningRateScheduler(scheduler)
```

### 1.1 parameter tuning for ResNet152V2 model

```python
model_name = "ResNet152V2"

fine_tune_list = [-10, -20, -40, -60]

activation_function = "relu"

learning_rate_list = [0.001, 0.004]

epoch = 20

output_units = len(class_names)
```

```python
%%time

## based on parameter tuning, get the best validation result for resnet model
best_val_accuracy_resnet = 0
best_model_resnet = None
train_acc_resnet = None
val_acc_resnet   = None
train_loss_resnet = None
val_loss_resnet = None

for fine_tune_at in fine_tune_list:
    for learning_rate in learning_rate_list:
        train_model = configure_model(model_name, fine_tune_at, output_units, activation_function, learning_rate)
        
        history = train_model.fit(train_dataset,
                            epochs=epoch,
                            batch_size = BATCH_SIZE,
                            validation_data=val_dataset,
                            callbacks = [st_callback, lr_callback])

        val_accuracy = max(history.history['val_accuracy'])

        ## update for the best result
        if val_accuracy > best_val_accuracy_resnet:
            best_val_accuracy_resnet = val_accuracy
            best_model_resnet = train_model
            train_acc_resnet  = history.history['accuracy']
            val_acc_resnet    = history.history['val_accuracy']
            train_loss_resnet = history.history['loss']
            val_loss_resnet   = history.history['val_loss']
            print()
            print("Up to now, the best result is {}, the layer is {}, the learning rate is {}".format(best_val_accuracy_resnet, fine_tune_at, learning_rate))
            print("------------------------------------------------------------------------------------")
            print()
```

```python
print("For ResNet model, the best result is {}, the layer is {}, the learning rate is {}".format(best_val_accuracy_resnet, fine_tune_at, learning_rate))
```

```python
loss0, accuracy0 = best_model_resnet.evaluate(val_dataset)
```

### 1.2 display learning curves

```python
acc = train_acc_resnet
val_acc = val_acc_resnet

loss = train_loss_resnet
val_loss = val_loss_resnet

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,1.0])
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```
